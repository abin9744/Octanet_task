The ATMs in our cities are built on Python, as we have all seen them. It is a console-based application with five different classes. In order to use the system, the user must enter his or her user ID and pin when it starts. Once the details are entered successfully, ATM functionality is unlocked. As a result of the project, the following operations can be performed:

Transactions History

Withdraw

Deposit

Transfer

Quit


```python
class User:
    def __init__(self, user_id, pin):
        self.user_id = user_id
        self.pin = pin
        self.balance = 0
        self.transaction_history = []

    def deposit(self, amount):
        self.balance += amount
        self.transaction_history.append(f"Deposited ${amount}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            self.transaction_history.append(f"Withdrew ${amount}")
        else:
            print("Insufficient funds!")

    def transfer(self, amount, recipient):
        if amount <= self.balance:
            self.balance -= amount
            recipient.balance += amount
            self.transaction_history.append(f"Transferred ${amount} to {recipient.user_id}")
            recipient.transaction_history.append(f"Received ${amount} from {self.user_id}")
        else:
            print("Insufficient funds!")

    def print_transaction_history(self):
        print("\nTransaction History:")
        for transaction in self.transaction_history:
            print(f"- {transaction}")

class ATM:
    def __init__(self):
        self.users = {}

    def create_user(self, user_id, pin):
        if user_id not in self.users:
            self.users[user_id] = User(user_id, pin)
            print("User created successfully!")
        else:
            print("User already exists!")

    def authenticate_user(self, user_id, pin):
        if user_id in self.users:
            if self.users[user_id].pin == pin:
                return self.users[user_id]
            else:
                print("Incorrect PIN!")
        else:
            print("User not found!")

def main():
    atm = ATM()
    
    while True:
        print("\n" + "=" * 30)
        print("Welcome to the ATM system!")
        print("=" * 30)
        print("1. Create User")
        print("2. Authenticate User")
        print("3. Quit")
        choice = input("Enter your choice: ")

        if choice == "1":
            user_id = input("Enter user ID: ")
            pin = input("Enter PIN: ")
            atm.create_user(user_id, pin)
        elif choice == "2":
            user_id = input("Enter user ID: ")
            pin = input("Enter PIN: ")
            user = atm.authenticate_user(user_id, pin)
            if user:
                while True:
                    print("\n" + "=" * 30)
                    print("ATM Operations:")
                    print("=" * 30)
                    print("1. Deposit")
                    print("2. Withdraw")
                    print("3. Transfer")
                    print("4. Transaction History")
                    print("5. Quit")
                    operation = input("Enter operation number: ")
                    
                    if operation == "1":
                        amount = float(input("Enter amount to deposit: "))
                        user.deposit(amount)
                        print("Deposit successful!")
                    elif operation == "2":
                        amount = float(input("Enter amount to withdraw: "))
                        user.withdraw(amount)
                    elif operation == "3":
                        recipient_id = input("Enter recipient's user ID: ")
                        amount = float(input("Enter amount to transfer: "))
                        recipient = atm.authenticate_user(recipient_id, "")
                        if recipient:
                            user.transfer(amount, recipient)
                            print("Transfer successful!")
                    elif operation == "4":
                        user.print_transaction_history()
                    elif operation == "5":
                        print("Returning to main menu...")
                        break
                    else:
                        print("Invalid operation!")
            else:
                print("Authentication failed!")
        elif choice == "3":
            print("Goodbye!")
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```
